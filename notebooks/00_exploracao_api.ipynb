```python
from nba_api.stats.endpoints import leaguedashplayerstats, playercareerstats
```

```python
df_season2023_24 = leaguedashplayerstats.LeagueDashPlayerStats(
    season='2023-24',
    season_type_all_star='Regular Season'
).get_data_frames()[0]

print(df_season2023_24.shape)
print(df_season2023_24.columns.tolist())
```

```python
df_min = df_season2023_24[df_season2023_24["TEAM_ABBREVIATION"]== 'MIN']
df_min.head()
```

```python
df_top10_s24 = df_season2023_24[df_season2023_24["PTS_RANK"] <= 10].sort_values(by='PTS_RANK',ascending=True)
df_top10_s24.head(10)
```
